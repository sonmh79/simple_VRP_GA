```python
def make_chromo():
  return random.sample(a,len(a))
```

```python
chromo = [make_chromo() for _ in range(100)]
chromo
```

```python
def distance(chromo):
  dist=0
  dist_list = []
  
  for route in chromo:
    r = route.copy()
    start = r[0]
    dist = 0

    while len(r)>1:
      dist += dataset.loc[r[0],r[1]]
      r.pop(0)
    dist += dataset.loc[r[0],start]
    dist_list.append(dist)
  fit = list(zip(dist_list,chromo))
  fit.sort(key = lambda x:x[0])

  return fit
```

```python
c_distance = distance(chromo)
print("최소 거리와 경로",c_distance[0])
print("최대 거리와 경로",c_distance[-1])
```

```python
class generation:
  cnt = 0
  def __init__(self,chromosomes,mutation_prob=0.2):
    generation.cnt +=1
    self.dna_list = chromosomes
    self.dna_len = len(chromosomes[0])
    self.dist_list = self.get_dist(self.dna_list) # result: [(distance1,[routes1]),...]
    self.fitness = np.mean([self.dist_list[i][0] for i in range(self.dna_len)])
    self.best_fit = self.dist_list[0]
    self.child = []
    self.mutation_prob = mutation_prob
  
  def get_dist(self, routes):
    dist = 0
    dist_list = []

    for route in routes:
      r = route.copy()
      start = r[0]
      dist = 0

      while len(r)>1:
        dist += dataset.loc[r[0],r[1]]
        r.pop(0)
      dist += dataset.loc[r[0],start]
      dist_list.append(dist)
    fit = list(zip(dist_list,routes))
    fit.sort(key = lambda x:x[0])

    return fit
  
  def rand(self,x,y):
    return int(random.uniform(x,y))

  def select_parent(self,parents):
    fitness = [100/route[0] for route in parents] 
    sum = 0

    for f in fitness:
      sum += f
    p_index = set()

    while len(p_index)<2:
      fit = 0
      target = random.uniform(0,sum)
      for i in range(len(fitness)):
        fit += fitness[i]
        if fit > target:
          p_index.add(i)
          break  

    return list(p_index)

  def make_child(self):
    parents = self.dist_list.copy()
    routes = [route[1] for route in parents]
    child = []

    while parents:
      selected_parents = self.select_parent(parents)
      p1,p2 = routes.pop(selected_parents[0]),routes.pop(selected_parents[1]-1)
      parents.pop(selected_parents[0]),parents.pop(selected_parents[1]-1)
      c1,c2 = self.crossover(p1,p2)
      child += c1,c2

    return child
  
  def crossover(self,p1,p2):
    swap_point = self.rand(1,self.dna_len-1)
    c1,c2 = [],[]

    i = 0
    while i < swap_point:
      c1.append(p1[i])
      c2.append(p2[i])
      i+=1

    for e in p2:
      if e not in c1:
        c1.append(e)

    for e in p1:
      if e not in c2:
        c2.append(e)
    
    #mutation
    if random.uniform(0,1) <= self.mutation_prob:
      e1,e2,r = random.sample(range(1,12),3)
      if r%2 == 0:
        c1[e1],c1[e2] = c1[e2],c1[e1]
      else:  
        c2[e1],c2[e2] = c2[e2],c2[e1]
    next_gen = self.get_dist([c1,c2,p1,p2])
    #print("next_gen: ",next_gen)
    c1,c2 = next_gen[0][1],next_gen[1][1]
    #print("c1,c2: ",c1,c2)
    return [c1,c2]

if __name__ == '__main__':
  cnt = 0
  mean_fitness = []
  best_fit = []
  first_generation = chromo
  n_generation = 100
  g = generation(first_generation)

  for i in range(n_generation):
    g = generation(g.make_child())
    cnt = g.cnt
    mean_fitness.append(g.fitness)
    best_fit.append(g.best_fit)

  best_fitness = [fit[0] for fit in best_fit]
  print("최단 거리: ",min(best_fit)[0])
  print("최단 루트: ",min(best_fit)[1])
  plt.plot(mean_fitness)
  plt.xlabel("n_generations")
  plt.ylabel("mean_fitness")
  plt.title("Mean distance per generation")
  plt.show()
  plt.plot(best_fitness)
  plt.xlabel("n_generations")
  plt.ylabel("best_fitness")
  plt.title("Minimun distance per generation")
  plt.show()
```
